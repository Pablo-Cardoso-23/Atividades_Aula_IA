```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
# Dados fictícios para regressão
dados = pd.DataFrame({
    'nota_prova1': [7.5, 4.0, 6.0, 8.5, 3.0, 9.0, 5.5, 2.0, 6.5, 7.0],
    'nota_prova2': [8.0, 5.0, 6.5, 9.0, 4.0, 9.5, 5.0, 3.0, 7.0, 6.0],
    'frequencia': [90, 60, 75, 95, 50, 98, 70, 40, 85, 80],
    'nota_final': [8.0, 5.0, 6.5, 9.0, 4.5, 9.5, 5.5, 3.0, 7.5, 7.0]
})
```

```python
# Separar variáveis independentes e alvo
X = dados[['nota_prova1', 'nota_prova2', 'frequencia']]
y = dados['nota_final']

# Dividir em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Criar e treinar o modelo
modelo = LinearRegression()
modelo.fit(X_train, y_train)
```

```python
# Fazer previsões
previsoes = modelo.predict(X_test)

# Calcular RMSE manualmente
rmse = np.sqrt(mean_squared_error(y_test, previsoes))
```

```python
# Gráfico de dispersão com linha de regressão
plt.figure(figsize=(8, 6))
x_plot = X_test.mean(axis=1)
plt.scatter(x_plot, y_test, color='blue', label='Real')
plt.scatter(x_plot, previsoes, color='red', label='Previsto')
plt.plot(np.sort(x_plot), previsoes[np.argsort(x_plot)], color='green', label='Linha de Regressão')
plt.xlabel('Média das variáveis explicativas')
plt.ylabel('Nota Final')
plt.legend()
plt.title('Regressão Linear - Previsão de Nota Final')
plt.tight_layout()
plt.savefig("grafico_regressao.png")
```

```python
# Exibir resultado
print("RMSE (Erro médio quadrático raiz):", round(rmse, 2))
```
