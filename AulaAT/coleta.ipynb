```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

```python
# Gerar dados fictícios de alunos
dados = pd.DataFrame({
    'nota_prova1': [7.5, 4.0, 6.0, 8.5, 3.0, 9.0, 5.5, 2.0, 6.5, 7.0],
    'nota_prova2': [8.0, 5.0, 6.5, 9.0, 4.0, 9.5, 5.0, 3.0, 7.0, 6.0],
    'frequencia': [90, 60, 75, 95, 50, 98, 70, 40, 85, 80],
    'situacao': ['Passou', 'Reprovou', 'Passou', 'Passou', 'Reprovou', 'Passou', 'Reprovou', 'Reprovou', 'Passou', 'Passou']
})
```

```python
# Separar variáveis independentes e alvo
X = dados[['nota_prova1', 'nota_prova2', 'frequencia']]
y = dados['situacao']
```

```python
# Dividir em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Criar e treinar o modelo
modelo = DecisionTreeClassifier(random_state=42)
modelo.fit(X_train, y_train)
```

```python
# Fazer previsões e calcular acurácia
previsoes = modelo.predict(X_test)
acuracia = accuracy_score(y_test, previsoes)
```

```python
# Plotar a árvore de decisão
plt.figure(figsize=(10,6))
plot_tree(modelo, feature_names=X.columns, class_names=modelo.classes_, filled=True)
plt.savefig("arvore_classificacao.png")

print("Acurácia do modelo de classificação:", round(acuracia * 100, 2), "%")
```
