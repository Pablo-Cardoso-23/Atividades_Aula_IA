## Explicação da Metodologia Escolhida

- Metodologia: Por envolver números e por se tratar de um sorteio, optei em usar a análise de padrões e estatística descritiva, usando o Pandas. Preferi não usar Machine Learning pois tende a errar mais e não tem como adivinhar o futuro, sendo assim eu achatei os dados de resultados e mapear a frequência absoluta de todos os números, o critério foi usar os números que mais aparecem.

- Justificativa: Na probabilidade, um sorteio é um evento independente, então minha ideia foi a de maximizar os meus resultados com base no número de aparições de cada número em N sorteios, como a base de dados é grande, creio que a análise com esse modelo me parece mais viável.

```python
# Instalar bibliotecas necessárias

!pip install pandas openpyxl --quiet

import pandas as pd
```

```python
dados = pd.read_excel('dados_historicos_sorteios.xlsx', skiprows=6)

display(dados)
```

```python
# Tratamento dos dados para pegar apenas as colunas importantes

colunas_uteis = ['bola 1', 'bola 2', 'bola 3', 'bola 4', 'bola 5', 'bola 6', 'bola 7', 'bola 8', 'bola 9', 'bola 10', 'bola 11', 'bola 12', 'bola 13', 'bola 14', 'bola 15']
bolas_sorteadas = dados[colunas_uteis]
```

```python
# Achatar o flaten
historico_numeros = bolas_sorteadas.values.flatten()

# Contar a frequência dos números
frequencia_numeros = pd.Series(historico_numeros).value_counts()

# Pegar os 15 números mais sorteados
numeros_mais_sorteados = frequencia_numeros.head(15)
```

```python
# Visualizar os resultados dos mais sorteados
print(f"Números Mais Frequentes | Quantidade de Vezes Sorteados: \n{numeros_mais_sorteados}")
```

```python
# Lista dos números para palpite
numeros_previstos = sorted(numeros_mais_sorteados.index.tolist())
print(f"\nLista dos 15 Números Previstos: {numeros_previstos}")
```
